```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
```

```python
# api 불러들이기
key = 'key'
rows = 3600 # 전체 정보 수 = 3522
url = f'http://openapi.kepco.co.kr/service/EvInfoServiceV2/getEvSearchList?serviceKey={key}&pageNo=1&numOfRows={rows}'
response = requests.get(url)
xml = BeautifulSoup(response.text, "html.parser")

# 자료 변환 (int->str)
chargeTp_dict = {"1":"완속", "2":"급속"}
cpStat_dict = {"0" : "상태확인불가", "1" : "충전가능", "2" : "충전중",
               "3" : "고장/점검", "4" : "통신장애", "9" : "충전예약", "5" : "알수없음"}
cpTp_dit = {"1" : "B타입(5핀)", "2" : "C타입(5핀)", "3" : "BC타입(5핀)", "4" : "BC타입(7핀)", 
            "5" : "DC차데모", "6" : "AC3상", "7" : "DC콤보", "8" : "DC차데모+DC콤보", 
            "9" : "DC차데모+AC3상", "10" : "DC차데모+DC콤보+AC3상", "13" : "알수없음"}

# 데이터 취합
data = []

for row in xml.find_all("item"):
    addr = row.find_all("addr")[0].text
    chargeTp = chargeTp_dict[row.find_all("chargetp")[0].text]
    cpid = row.find_all("cpid")[0].text
    cpnm = row.find_all("cpnm")[0].text
    cpstat = cpStat_dict[(row.find_all("cpstat")[0].text)]
    cptp = cpTp_dit[row.find_all("cptp")[0].text]
    csid = row.find_all("csid")[0].text
    csnm = row.find_all("csnm")[0].text
    lat = row.find_all("lat")[0].text
    long = row.find_all("longi")[0].text
    statupdatedatetime = row.find_all("statupdatedatetime")[0].text

    data.append([addr, chargeTp, cpid, cpnm, cpstat, cptp, csid, csnm, lat, long, statupdatedatetime])

df = pd.DataFrame(data, columns=["충전소주소", "충전기타입", "충전기ID", "충전기명칭",
                            "충전기 상태 코드", "충전방식", "충전소ID", "충전소 명칭", 
                            "위도", "경도", "충전기 상태 갱신 시각"])
df
```

```python
df.to_csv("./data/충전소_정보.csv", index=False)
```
